#Scrap

```python
urls = ['https://us.spindices.com/indices/equity/sp-bmv-construction-index',
        'https://us.spindices.com/indices/equity/sp-bmv-housing-index',
        'https://us.spindices.com/indices/equity/sp-bmv-mining-agriculture-index',
        'https://us.spindices.com/indices/equity/sp-bmv-telecommunication-services-sector-index',
        'https://us.spindices.com/indices/equity/sp-bmv-financial-services-index'
        ]
```

```python
def price_return(url):
    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless=False)
    browser.visit(url)
    time.sleep(10)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    price = soup.find('td',class_='price-return')
    browser.quit()
    return price
```

```python
index_v = []
for x in urls:
    index_v.append(price_return(x).text)
```

```python
today = dt.today()
```

```python
new_entry = {
        'Date':today,
        'Construction Index':float(index_v[0]),
        'Housing Index':float(index_v[1]),
        'Mining & Agriculture Index':float(index_v[2]),
        'Telecommunication Services Sector Index':float(index_v[3]),
        'Financial Services Index':float(index_v[4])
    }

collection.insert_one(new_entry)
```
